# **ANN - Binary Problem**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.diagnosis.value_counts()
```

```python
df.diagnosis.replace({'B':0,'M':1},inplace=True)
```

```python
df.columns
```

```python
y = df['diagnosis']
X = df.drop(['id', 'diagnosis','Unnamed: 32'],axis=1)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y,random_state=2529)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
ann = Sequential()

ann.add(Dense( 30, activation = 'relu'))
ann.add(Dense(60, activation = 'relu'))
ann.add(Dense(20, activation = 'relu'))
ann.add(Dense(1, activation = 'sigmoid'))

ann.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size=32,epochs=20)
```

```python
ann.summary()
```

```python
y_pred = ann.predict(X_test)
y_pred = y_pred.round()
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# **Multiclass Categorical Problem**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/IRIS.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
y = df['species']
X =df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
```

```python
y
```

```python
y = pd.get_dummies(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=y,random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann = Sequential()
ann.add(Dense(4, activation='relu'))
ann.add(Dense(10, activation='relu'))
ann.add(Dense(10, activation='relu'))
ann.add(Dense(3, activation='softmax'))

ann.compile(optimizer='adam',loss = 'categorical_crossentropy',metrics=['accuracy'])

ann.fit(X_train,y_train,batch_size=4,epochs=30)
```

```python
ann.summary()
```

Iris-setosa	Iris-versicolor	Iris-virginica

```python
y_pred = ann.predict(X_test)
```

```python
y_pred = np.argmax(y_pred,axis=1)
```

```python
y_test = np.argmax(y_test, axis=1)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

# **Convolution NeuralNetwork**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
```

```python
# read inbuild dataset
from tensorflow.keras.datasets import fashion_mnist
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
plt.imshow(X_train[3])
```

```python
X_train.shape
```

```python
X_train = X_train.reshape(60000, 28, 28, 1)
```

```python
X_test = X_test.reshape(10000, 28, 28, 1)
```

```python
# scale
X_train.min(), X_train.max()
```

```python
X_train = X_train/255
```

```python
X_test = X_test/255
```

```python
y_train
```

```python
np.unique(y_train)
```

```python
y_train_dum = tf.keras.utils.to_categorical(y_train)
```

```python
y_test_dum = tf.keras.utils.to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten
```

```python
model = Sequential()

model.add(Conv2D(filters=64, kernel_size= (3,3), input_shape = (28,28,1), activation='relu', ))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(20, activation='relu'   ))
model.add(Dense(15, activation='relu'   ))
model.add(Dense(10, activation='softmax'   ))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train_dum,epochs=2)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.argmax(y_pred,axis=1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
